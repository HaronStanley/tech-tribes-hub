```python
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
population_kenya = pd.read_csv("population_kenya.csv")
```

```python
# Display the first few rows of the population_kenya DataFrame
population_kenya.head()
```

```python
# Get the shape of the population_kenya DataFrame
population_kenya.shape
```

```python
# Generate descriptive statistics for the population data of Kenya
population_kenya.describe()
```

```python
# Display information about the population_kenya dataset
population_kenya.info()
```

```python
# Plotting a pie chart
import matplotlib.pyplot as plt
labels = ['Females', 'Males']
sizes = [population_kenya['Females'].sum(), population_kenya['Males'].sum()]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution in Kenya')
plt.axis('equal')
plt.show()

```

```python
# Plotting a bar graph
import matplotlib.pyplot as plt

labels = ['Females', 'Males']
sizes = [population_kenya['Females'].sum(), population_kenya['Males'].sum()]
plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=['blue', 'green'])
plt.xlabel('Gender')
plt.ylabel('Population')
plt.title('Gender Distribution in Kenya')
plt.show()

```

```python
# Plotting a cumulative bar graph
import matplotlib.pyplot as plt
labels = ['Females', 'Males']
sizes = [population_kenya['Females'].sum(), population_kenya['Males'].sum()]
cumulative_sizes = [sum(sizes[:i+1]) for i in range(len(sizes))]
plt.figure(figsize=(8, 6))
plt.bar(labels, cumulative_sizes, color=['blue', 'green'])
plt.xlabel('Gender')
plt.ylabel('Cumulative Population')
plt.title('Cumulative Gender Distribution in Kenya')
plt.show()
```
